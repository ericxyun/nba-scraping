```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

pd.set_option('display.max_columns', 500)
```

```python
matplotlib.rc('axes', titlesize=20)
matplotlib.rc('xtick', labelsize=20)     
matplotlib.rc('ytick', labelsize=20)
matplotlib.rc('axes', labelsize=20)
```

```python
data = pd.read_csv('data/total/2019-01-10_total.csv')
data.head()
```

```python
df = pd.DataFrame(data.AGE.value_counts(sort=True)).reset_index()
df.rename(columns={'index': 'Age', 'AGE':'count'}, inplace=True)
df.sort_values(by='Age').plot('Age', 'count', kind='bar', colormap='Set3', figsize=(15,5), title='NBA Player Age', legend=False)
plt.xticks(rotation=45)
plt.xlabel('Age', size=20)
plt.ylabel('Count', size=20)
matplotlib.rc('axes', titlesize=20)
matplotlib.rc('xtick', labelsize=20)     
matplotlib.rc('ytick', labelsize=20)
display(data[['PLAYER', 'AGE']].sort_values(by='AGE', ascending=False).head())
display(data[['PLAYER', 'AGE']].sort_values(by='AGE', ascending=False).tail())
plt.show()
```

__Games Played__

```python
df = pd.DataFrame(data.GP.value_counts(sort=True)).reset_index()
df.rename(columns={'index': 'GP', 'GP':'count'}, inplace=True)
df.plot('GP', 'count', kind='bar', colormap='Set3', figsize=(20,5), title='NBA Player GP', legend=False)
plt.xticks(rotation=45)
plt.xlabel('GP', size=20)
plt.ylabel('Count', size=20)
matplotlib.rc('axes', titlesize=20)
matplotlib.rc('xtick', labelsize=10)     
matplotlib.rc('ytick', labelsize=10)
display(data[['PLAYER', 'GP']].sort_values(by='GP', ascending=False).head())
display(data[['PLAYER', 'GP']].sort_values(by='GP', ascending=False).tail())
plt.show()
```

__Field Goal Attempt/ Free Throw Attempt__  
As NBA players shoot more, they are more prone to drawing fouls

```python
display(data[['PLAYER', 'TEAM', 'FGA', 'FTA']].head(10).sort_values(by='FTA', ascending=False))
data[['FGA','FTA']].plot(x='FGA', y='FTA', kind='scatter', title='Field Goal Average vs Free Throw Attempt', figsize=(10,8))
```

__Field Goal Attempts vs. Turnover Rate__ 
Some of the best NBA players have the highest turn over rates, mainly because they handling the bar more.

```python
display(data[['PLAYER', 'TEAM', 'FGA', 'TOV']].head(10).sort_values(by='TOV', ascending=False))
data[['FGA','TOV']].plot(x='FGA', y='TOV', kind='scatter', figsize=(10,8))
```

__Field Goal Average vs Steals__  
Better shooters are better stealers

```python
display(data[['PLAYER', 'TEAM', 'FGA', 'STL']].head(10).sort_values(by='STL', ascending=False))
data[['FGA','STL']].plot(x='FGA', y='STL', kind='scatter', figsize=(10,8))
```

__Field Goal Average vs Personal Fouls__  
Better shooters are also more agressive players?

```python
display(data[['PLAYER', 'TEAM', 'FGA', 'PF']].head(10).sort_values(by='PF', ascending=False))
data[['FGA','PF']].plot(x='FGA', y='PF', kind='scatter', figsize=(10,8))
```

```python
display(data[['PLAYER', 'TEAM', 'AGE', 'FG%']].head(10).sort_values(by='FG%', ascending=False))
data[['FG%','AGE']].plot(x='AGE', y='FG%', kind='scatter', figsize=(10,8))
```

```python
def correlation(column):
    error_catch = []
    for item in data.drop(column, axis=1).columns.tolist():
        try:

            data[[column, item]].plot(x=column, y=item, kind='scatter', figsize=(8,6))
            plt.show()
            display(data[['PLAYER', 'TEAM', column, item]].head(10).sort_values(by=item, ascending=False))
        except:
            error_catch.append(str(item) + " couldn't be analyzed")

    for item in error_catch:
        print(item)
```
